```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd

base_de_dados = pd.read_excel('dados_estudantes.xlsx')

base_de_dados
```

```python
data = base_de_dados['Turma'].value_counts().reset_index()

print(data)

labels = data['Turma']
tamanhos = data['count']

plt.pie(tamanhos, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Alunos por turma')
plt.axis('equal')
plt.show()
```

```python
data = base_de_dados['Sexo'].value_counts().reset_index()

print(data)

labels = data['Sexo']
tamanhos = data['count']

plt.pie(tamanhos, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Genero de Alunos')
plt.axis('equal')
plt.show()
```

```python
data = base_de_dados['Idade'].value_counts().reset_index()

print(data)

objects = data['Idade']
performance = data['count']

y_pos = np.arange(len(objects))

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Recorrencia')
plt.title('Idade Alunos')

plt.show()
```
