```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('.\data\pobreza_brasil.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
values = df['Genero'].value_counts(normalize=True)
plt.pie(values, labels=(values.index), autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Pastel1'))
plt.title('Divisão ddos dados por gênero')
plt.show()
```

```python
values = df['Idade'].value_counts()
color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.6 # transparency
         ) 
plt.bar(x=(values.index), height=values, align='center', alpha=0.5, color=color)
plt.title('Divisão dos dados por idade')
plt.show()
```
