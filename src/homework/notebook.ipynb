```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('data/pobreza_brasil.xlsx')
df.head()
color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.6 # transparency
         ) 
```

```python
df.info()
```

```python
df.describe()
```

```python
values = df['Genero'].value_counts(normalize=True).reset_index()

print(values)
plt.pie(values, labels=(values.index), autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Pastel1'))
plt.title('Divisão dos dados por gênero')


# labels = df['OpTV']
# tamanhos = df['count']

plt.show()
```

```python
values = df['Idade'].value_counts()
plt.bar(x=(values.index), height=values, align='center', alpha=0.5, color=color)
plt.title('Divisão dos dados por idade')
plt.show()
```

```python
values = df['Escolaridade'].value_counts()
plt.bar(x=(values.index), height=values, align='center', alpha=0.5, color=color)
plt.title('Divisâo dos dados por escolaridade')
plt.show()
```

```python
values = df['Trabalho'].value_counts()
plt.bar(x=values.index, height=values, align='center', alpha=0.5, color=color)
plt.title('categorias de trabalho')
plt.show()
```

```python
values = df['Reg_metrop'].value_counts()
plt.pie(values, labels=(values.index), autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Pastel1'))
plt.title('Divisão dos dados por região')
plt.show()
```

```python
values = df['Cor'].value_counts()
plt.pie(values, labels=(values.index), autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Pastel1'))
plt.title('Divisão dos dados por cor de pele')
plt.show()
```

```python
values = df['Cart_assinada'].value_counts()

plt.bar(x=values.index, height=values, align='center', alpha=0.5, color=color)
plt.title('Divisão dos dados por carteira de trabalho assinada')
plt.show()
```

```python
values = df['Pobreza'].value_counts()
plt.pie(values, labels=(values.index), autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Pastel1'))
plt.title('Divisão dos dados por pobreza')
plt.show()
```

```python
# Criando uma tabela de frequências cruzadas
cross_tab = pd.crosstab(df['Trabalho'], df['Genero'])

# Plotando o gráfico de barras agrupadas
cross_tab.plot(kind='barh', stacked=False,color=sns.color_palette('Pastel1'))

plt.title('Tipos de Trabalho por Gênero')
plt.xlabel('Quantidade de Pessoas')
plt.ylabel('Tipo de Trabalho')

plt.show()
```

```python
# Criando uma tabela de frequências cruzadas
cross_tab = pd.crosstab(df['Escolaridade'], df['Cor'])

# Plotando o gráfico de barras empilhadas
cross_tab.plot(kind='bar', stacked=True,color= sns.color_palette('Pastel1'))

plt.title('Escolaridade por Cor')
plt.xlabel('Escolaridade')
plt.ylabel('Quantidade de Pessoas')


plt.show()
```

```python
sns.countplot(x='Cart_assinada', hue='Pobreza', data=df)
plt.title('Relação entre Carteira Assinada e Pobreza')
plt.xlabel('Carteira Assinada')
plt.ylabel('Contagem')
plt.show()
```
