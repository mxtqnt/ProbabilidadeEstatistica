```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd

base_de_dados = pd.read_excel('data/pobreza_brasil.xlsx')

base_de_dados
```

```python
# Pizza Pobreza

data = base_de_dados['Pobreza'].value_counts().reset_index()

labels = data['Pobreza']
tamanhos = data['count']

plt.pie(tamanhos, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('RelaÃ§Ã£o Pobreza')
plt.axis('equal')
plt.show()
```

```python
# Colunas Pobreza x Genero

data = base_de_dados['Genero'].value_counts().reset_index()

print(data)

labels = data['Pobreza']
tamanhos = data['count']

plt.pie(tamanhos, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('RelaÃ§Ã£o Pobreza x Genero')
plt.axis('equal')
plt.show()
```
