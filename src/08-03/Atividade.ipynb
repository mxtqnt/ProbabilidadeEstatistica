```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
import seaborn as sns 

base_de_dados = pd.read_excel('data/dados_estudantes.xlsx')

print(base_de_dados)

color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.6 # transparency
         ) 
```

```python
data = base_de_dados['Turma'].value_counts().reset_index()

labels = data['Turma']
tamanhos = data['count']

plt.pie(tamanhos, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Pastel1'))
plt.title('Alunos por turma')
plt.axis('equal')

data = data.rename(columns={"count": "Contagem"})
data = data.style.set_caption("Alunos por turma")

display(data)

plt.show()
```

```python
data = base_de_dados['Idade'].value_counts().reset_index()

objects = data['Idade']
performance = data['count']

y_pos = np.arange(len(objects))

plt.bar(y_pos, performance, align='center', alpha=0.5, color=color)
plt.xticks(y_pos, objects)
plt.ylabel('Recorrencia')
plt.title('Idade Alunos')

data = data.rename(columns={"count": "Contagem"})
data = data.style.set_caption("Idade Alunos")

display(data)

plt.show()
```

```python
data = base_de_dados[['Fuma', 'Sexo']]

# Contagem de ocorrências para cada combinação de Fuma e Sexo
count_data = data.groupby(['Fuma', 'Sexo']).size().unstack()
# Plotando o gráfico de barras empilhadas
count_data.plot(kind='bar', stacked=False, color=sns.color_palette('Pastel1'))

plt.title('Relação entre Fumante e Gênero')
plt.xlabel('Fuma')
plt.ylabel('Contagem')
plt.legend(title='Sexo')

plt.show()
```

```python
data = base_de_dados[['Fuma', 'Toler']]

# Contagem de ocorrências para cada combinação de Fuma e Toler
count_data = data.groupby(['Fuma', 'Toler']).size().unstack()

# Definindo cores
colors = sns.color_palette('Pastel1')

# Plotando o gráfico de barras empilhadas
ax = count_data.plot(kind='bar', stacked=False, color=colors)

# Editando legendas
legend_labels = ['Indiferente', 'Pouco Incômodo', 'Muito Incômodo']
ax.legend(legend_labels, title='Tipo de Tolerância')

plt.title('Relação entre Tolerância e Fumante')
plt.xlabel('Fuma?')
plt.ylabel('Contagem')

plt.show()
```
